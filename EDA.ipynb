### Basic data manipulation

#### Extract the subset of data for one particular country region. 

To understand the overall trend and mobility, I would use the subset of data where country_region is equal to “United States” and sub_region_1 is null

```python
data_subset = data[(data['country_region'].eq("United States")) & (data['sub_region_1'].isnull())]
```

### Data Visualization

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=5, cols=1)

fig.add_trace(go.Scatter( x= data_subset["date"], y=data_subset["grocery_and_pharmacy_percent_change_from_baseline"], 
                         name = 'grocery_and_pharmacy_percent_change_from_baseline'),
              row = 1, col = 1
)

fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["parks_percent_change_from_baseline"], 
               name = 'parks_percent_change_from_baseline'),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["transit_stations_percent_change_from_baseline"], 
               name = 'transit_stations_percent_change_from_baseline'),
    row=3, col=1
)


fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["workplaces_percent_change_from_baseline"], 
               name = 'workplaces_percent_change_from_baseline' ),
    row=4, col=1
)

fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["residential_percent_change_from_baseline"], 
              name = 'residential_percent_change_from_baseline'),
    row=5, col=1
)

fig.update_layout(height=1700, width=1200)
fig.show()
```

### Did anything significant happen between December 1, 2020 and March 1, 2021?

1. Did united states region experience a cold winter or a summer transitioning into fall?
2. Did public holidays fall on the same day of the week?

```python
df = data_subset[data_subset["date"].between('2020-12-01', '2021-03-01')]
```

```python

fig = make_subplots(rows=5, cols=1)

fig.add_trace(go.Scatter( x= df["date"], y=df["grocery_and_pharmacy_percent_change_from_baseline"], 
                         name = 'grocery_and_pharmacy_percent_change_from_baseline'),
              row = 1, col = 1
)

fig.add_trace(
    go.Scatter(x=df["date"], y=df["parks_percent_change_from_baseline"], 
               name = 'parks_percent_change_from_baseline'),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=df["date"], y=df["transit_stations_percent_change_from_baseline"], 
               name = 'transit_stations_percent_change_from_baseline'),
    row=3, col=1
)


fig.add_trace(
    go.Scatter(x=df["date"], y=df["workplaces_percent_change_from_baseline"], 
               name = 'workplaces_percent_change_from_baseline' ),
    row=4, col=1
)

fig.add_trace(
    go.Scatter(x=df["date"], y=df["residential_percent_change_from_baseline"], 
              name = 'residential_percent_change_from_baseline'),
    row=5, col=1
)

fig.update_layout(height=1700, width=1200)
fig.show()
```
